## Getting things ready

First, import the libraries that you are going to need. You could also do
that later, but it is better style to do that at the beginning.
pylab imports the numpy, scipy, and matplotlib.pyplot libraries into the
current environment

```python
%pylab inline

import scipy.stats as stats
import seaborn as sns
sns.set_context('notebook')
sns.set_style('darkgrid')
```

```python
# Generate data that are normally distributed
x = randn(1000)
```

## Scatter plot

```python
plot(x,'.')
title('Scatter Plot')
xlabel('X')
ylabel('Y')
draw()
```

## Histogram

```python
hist(x)
xlabel('Data Values')
ylabel('Frequency')
title('Histogram, default settings')
```

```python
hist(x,25)
xlabel('Data Values')
ylabel('Frequency')
title('Histogram, 25 bins')
```

### KDE

```python
import seaborn as sns
sns.kdeplot(x)
xlabel('Data Values')
ylabel('Density')
```

## Cumulative probability density

```python
numbins = 20
cdf = stats.cumfreq(x,numbins)
plot(cdf[0])
xlabel('Data Values')
ylabel('Cumulative Frequency')
title('Cumulative probablity density function')
```

## Boxplot

```python
# The error bars indicate 1.5* the inter-quartile-range (IQR), and the box consists of the
# first, second (middle) and third quartile
boxplot(x, sym='*')
title('Boxplot')
ylabel('Values')
```

```python
boxplot(x, vert=False)
title('Boxplot, horizontal')
xlabel('Values')
```

## Errorbars

```python
x = arange(5)
y = x**2
errorBar = x/2
errorbar(x,y, yerr=errorBar, fmt='o', capsize=5, capthick=3)

plt.xlabel('Data Values')
plt.ylabel('Measurements')
plt.title('Errorbars')

xlim([-0.2, 4.2])
ylim([-0.2, 19])
```

## Check for Normality

```python
# Visual check
x = randn(100)
_ = stats.probplot(x, plot=plt)
title('Probplot - check for normality')
```

## 2D Plot
